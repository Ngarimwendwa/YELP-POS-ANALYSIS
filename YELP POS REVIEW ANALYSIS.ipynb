```python
DATA_PATH = "dataset.json"  # Do not rename

def load_json_dict_of_dicts(path: str) -> pd.DataFrame:
    with open(path, "r", encoding="utf-8") as f:
        data = json.load(f)


    if not isinstance(data, dict) or not all(isinstance(v, dict) for v in data.values()):
        raise ValueError("Expected dict-of-dicts JSON structure.")

    df = pd.DataFrame(data).reset_index(drop=True)
    return df

df = load_json_dict_of_dicts(DATA_PATH)
df.head()
```

```python
#Converting the miliseconds date on the furthest left into  datetime stamp
df["date"] = pd.to_datetime(df["date"], unit="ms", errors="coerce")
df["year"] = df["date"].dt.year

# Filter years 2017–2022 inclusive
df = df[(df["year"] >= 2017) & (df["year"] <= 2022)].copy()

df[["date", "year"]].head(), df["year"].value_counts().sort_index()

Original_df = df.copy()
```

```python
stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()  # optional if you choose stemming

def normalize_text(text: str) -> str:
    """
        - Lowercase
        - Remove numbers
        - Remove punctuation/special chars
        - Remove extra whitespace
    """
    if pd.isna(text):
        return ""
    text = text.lower()
    text = re.sub(r"\d+", " ", text)              # remove numbers
    text = re.sub(r"[^a-z\s]", " ", text)         # keep letters + whitespace only
    text = re.sub(r"\s+", " ", text).strip()      # collapse whitespace
    return text

def tokenize(text: str) -> list:
    return nltk.word_tokenize(text)

def remove_stopwords(tokens: list) -> list:
    return [t for t in tokens if t not in stop_words and len(t) > 1]

def lemmatize_tokens(tokens: list) -> list:
    # Simple lemmatization (without POS-based lemmatization)
    return [lemmatizer.lemmatize(t) for t in tokens]

def stem_tokens(tokens: list) -> list:
    return [stemmer.stem(t) for t in tokens]
```

```python
# Option 1:
# Defining a sample rate, in this case 200.000, compared to 3.8 million
SAMPLE_SIZE = 200000

# Defining a random seed --->> for experiment reproduction
RANDOM_SEED = 42

# Creating a small sample out of the big DB.
df_small = df.sample(SAMPLE_SIZE, random_state=RANDOM_SEED).copy()
print(len(df_small))
```

```python
# Apply preprocessing pipeline

#1. Normalize text
df_small["text_norm"] = df_small["text"].apply(normalize_text)

#2. Tokenize the Text
df_small["tokens"] = df_small["text_norm"].apply(tokenize)

#3. Remove stop words
df_small["tokens_nostop"] = df_small["tokens"].apply(remove_stopwords)

#4. Lemmatize (Preferred over Stemming for interpretability)
df_small["tokens_lemmatized"] = df_small["tokens_nostop"].apply(lemmatize_tokens)

#5. Join tokens back into a string (This is appropriate if TD-IDF is put under consideration)
df_small["text_processed"] = df_small["tokens_lemmatized"].apply(lambda t: " ".join(t))

# Display the results to see if the code works
df_small[["text", "text_norm", "text_processed"]].head()
```

```python
# Task 2.1: Perform POS tagging

def get_pos_tags(tokens_list):
    """
    This takes a list of tokens, runs nltk.pos_tag, 
    and returns just a list of the POS tags.
    """
    # nltk.pos_tag returns tuples like: [('great', 'JJ'), ('food', 'NN')]
    tagged_tuples = pos_tag(tokens_list)
    
    # Extract just the tags (the second item in the tuple)
    return [tag for word, tag in tagged_tuples]

print("Starting POS Tagging... (This may take a minute)")
df_small['pos_tags'] = df_small['tokens_lemmatized'].apply(get_pos_tags)
print("Finished POS Tagging!")


#Task 2.2: Visualize the distribution

# Flatten the list of lists into one giant list containing all tags from all reviews
all_tags = [tag for tags_list in df_small['pos_tags'] for tag in tags_list]

# Count how many times each tag appears
tag_counts = Counter(all_tags)

# Create a DataFrame of the top 20 most common tags for our chart
tag_df = pd.DataFrame.from_dict(tag_counts, orient='index', columns=['Count']).reset_index()
tag_df = tag_df.rename(columns={'index': 'POS_Tag'})
tag_df = tag_df.sort_values(by='Count', ascending=False)
top_20_tags = tag_df.head(20)

# Build the Bar Chart using matplotlib
plt.figure(figsize=(12, 6))
plt.bar(top_20_tags['POS_Tag'], top_20_tags['Count'], color='coral')
plt.title('Distribution of Top 20 POS Tags in Yelp Reviews')
plt.xlabel('Parts of Speech (POS) Tag')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Task 3.1: Extract named entities

# We only want to look for specific types of entities relevant to Yelp reviews
# ORG: Companies, agencies, institutions (Businesses)
# PRODUCT: Objects, vehicles, foods, etc. (Not services)
# GPE: Countries, cities, states (Locations)
# LOC: Non-GPE locations, mountain ranges, bodies of water
TARGET_ENTITIES = {'ORG', 'PRODUCT', 'GPE', 'LOC'}

def extract_entities(text):
    """
    Runs spaCy NLP on the original text and extracts 
    entities that match our target list.
    """
    # Check for empty text
    if pd.isna(text) or not str(text).strip():
        return []
    
    # Process text with spaCy (limit length if texts are huge, but Yelp reviews are fine)
    doc = nlp(str(text))
    
    # Return a list of the text of the entity (lowercased to group them better) 
    # and its label if it's in our target list
    entities = [ent.text.lower().strip() for ent in doc.ents if ent.label_ in TARGET_ENTITIES]
    return entities

print("Starting Named Entity Extraction... (This is computationally heavy and will take time!)")
# Tip: If your df_small is 200,000 rows, this might take 15-30 minutes depending on your computer.
# If it's taking too long for testing, try running it on just: df_small.head(1000)['text'].apply(...) first!
df_small['named_entities'] = df_small['text'].apply(extract_entities)
print("Finished Entity Extraction!")


# Task 3.2: Summarize the top 10 most frequently mentioned entities 

# Flatten the list of lists into one giant list of all entities
all_entities = [entity for entity_list in df_small['named_entities'] for entity in entity_list]

# Filter out empty strings just in case
all_entities = [e for e in all_entities if e]

# Count how many times each entity appears
entity_counts = Counter(all_entities)

# Get the top 10
top_10_entities = entity_counts.most_common(10)

print("\n--- Top 10 Most Frequently Mentioned Entities ---")
for i, (entity, count) in enumerate(top_10_entities, 1):
    print(f"{i}. {entity.title()} ({count} mentions)")

# Optional: Visualize the top 10 Entities
entity_df = pd.DataFrame(top_10_entities, columns=['Entity', 'Count'])
plt.figure(figsize=(10, 5))
plt.barh(entity_df['Entity'][::-1], entity_df['Count'][::-1], color='skyblue') # [::-1] puts the highest at the top
plt.title('Top 10 Named Entities in Yelp Reviews')
plt.xlabel('Frequency')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
#Task 4.1: Represent the text numerically

# Method A: TF-IDF

print("Building TF-IDF model...")

# Initialize the vectorizer. We limit max_features so it doesn't crash our memory!
# We also use English stop words just in case any slipped through
tfidf = TfidfVectorizer(max_features=1000, stop_words='english')

# We fit it on our final processed text column (which is a string of words)
tfidf_matrix = tfidf.fit_transform(df_small['text_processed'])

# Print out the shape to show the user what it built
print(f"TF-IDF Matrix Shape: {tfidf_matrix.shape}")
print(f"This means {tfidf_matrix.shape[0]} reviews are each represented by {tfidf_matrix.shape[1]} numeric features.")


# ------------------------------------------------------------------------------------------------------------------------------------------------
# Method B: Word2Vec
# --------------------------
print("\nBuilding Word2Vec model...")

# Word2Vec expects a list of lists of tokens (not a full string)
# e.g., [['good', 'food'], ['bad', 'service']]
token_lists = df_small['tokens_lemmatized'].tolist()

# Train the Word2Vec model
# vector_size: How many dimensions to represent each word in
# window: How many words around the target word to look at for context
# min_count: Ignore words entirely if they appear less than 5 times
w2v_model = Word2Vec(sentences=token_lists, vector_size=100, window=5, min_count=5, workers=4)

print("Word2Vec model trained successfully!")

# Let's test the Word2Vec model to establish semantic understanding!
test_word = "pizza"
if test_word in w2v_model.wv:
    print(f"\nWords most similar to '{test_word}' according to Word2Vec:")
    similar_words = w2v_model.wv.most_similar(test_word, topn=5)
    for word, similarity in similar_words:
        print(f"  - {word}: {similarity:.3f}")
else:
    print(f"'{test_word}' wasn't found in the vocabulary.")
```
